```python
def tacgiaLink():
  global author_name,list_author,bod,list_quote
  author_name = [] # tên tác giả
  list_author = []  # nằm trong phần link của các tác giả
  bod = [] #birthday
  list_quote = []
  for link in links:
    r = requests.get(link)
    soup = BeautifulSoup(r.content,'html.parser')
    result = soup.find_all('div',attrs={'class':'quote'})
    # print(result)
    for item in result:
      c = item.find_all('small',attrs={'class':'author'})
      # tên tác giả
      for name in c:
          test = name.text
          author_name.append(test)
          auth_name = [i for n, i in enumerate(author_name) if i not in author_name[:n]] # tên tác giả sau khi lọc các phần tử trùng
      quotes = item.find_all('span',attrs={'class':'text'})
      for quote in quotes:
          q = quote.text
          list_quote.append(q)
      #link tác giả
      for url in item.find_all('a',href=True,text='(about)'):
          a = 'http://quotes.toscrape.com'
          b = a + url['href']
          list_author.append(b) 
          link_author = [i for n, i in enumerate(list_author) if i not in list_author[:n]] # link của các tác giả sau khi lọc các phần tử trùng
  # print(link_author)
  #ngày tháng năm sinh 
  for years in list_author:
        age = requests.get(years)
        soup_age = BeautifulSoup(age.content,'html.parser')
        birthday = soup_age.find_all('span',attrs={'class':'author-born-date'})
        for k in birthday:
          if(k != ''):
            bod.append(k.text)
          else:
            bod.append('Unknown')
  # #Đưa dữ liệu vào file csv 
  from google.colab import drive
  from nltk.metrics import edit_distance
  import csv
  testing = []
  for i, j, k, z in zip(author_name,list_author,bod, list_quote):
        testing.append((i,j,k,z)) 
  # print(testing)
  df = pd.DataFrame(testing,columns=['Tác giả', 'Link tác giả', 'Ngày tháng năm sinh', 'Quote'])
  drive.mount('drive',force_remount=True) 
  df.to_csv('Quote.csv',encoding='utf-8')
  b = df.head(10)
  return b
```
